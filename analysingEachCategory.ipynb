```python
temp = 0
frames = []
for i in range(16):
    increaseCount = 0 
    decreaseCount = 0
    path = f"C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\16categories\\category_number_{i}.csv"
    df = pd.read_csv(path)
    totalCount = len(df)
    temp += totalCount
    for j in range(totalCount):
        if(df['Change'][j] < 0):
            decreaseCount += 1
        else:
            increaseCount += 1
    avg = df['Change'].mean()
    data = {
        "Category" : i, 
        "Decrease Percentage" : f"{round((decreaseCount*100)/totalCount,2)}%",
        "Increase Percentage": f"{round((increaseCount*100)/totalCount,2)}%", 
        "Average Change(%)": f"{round(avg*100,2)}%"
    }
    temp_df = pd.DataFrame(data, index=[1])
    frames.append(temp_df)
    
    print(f"Catergory {i} :")
    print(f"Total Count : {totalCount}")
    print(f"Decrease Count : {decreaseCount} , Decrease Percentage : {round((decreaseCount*100)/totalCount,2)}% ")
    print(f"Total Count : {totalCount} , Increase Percentage : {round((increaseCount*100)/totalCount,2)}%")
    print(f"Average Change(%) : {round(avg*100,2)}%\n\n")
    
```

```python
df = pd.concat(frames, ignore_index = False)
```
