```python
import pandas as pd
from os import listdir
```

```python
data_df = pd.read_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\modelling\\dataset\\trainDataset.csv")
sector_df = pd.read_csv('C:\\Users\\garya\\ML projects\\Stock analysis\\sectorwiseanalysis\\dataset\\sectorIndex.csv').set_index('SYMBOL')
```

```python
sectors = listdir("C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\nifty indexes")
```

```python
sectors

```

```python
frames = []
for sector in sectors: 
    for i in range(16):
        data = {
            "Category" : i , 
            "Sector": sector[4:-4], 
            "Total Count" : 0, 
            "Increase Count" : 0, 
            "Decrease Count" : 0, 
            "Increase Percentage": 0, 
            "Decrease Percentage" : 0, 
            "Average Change" : 0
        }
        temp_df = pd.DataFrame(data, index = [1])
        frames.append(temp_df)
        

df = pd.concat(frames, ignore_index = True)
```

```python
df
```

```python
x = 0
for sector in sectors:
    path = f"C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\nifty indexes\\{sector}"
    temp_df = pd.read_csv(path)
    for i in range(len(temp_df)):
        symbol = temp_df['Symbol'].loc[i]
        sym_df = data_df[data_df['SYMBOL'] == symbol]
        
        for j in range(len(sym_df)):
            category = sym_df['Category'].iloc[j]
            print(category)
            print(sym_df['SYMBOL'].iloc[j])
            print(sym_df['Change'].iloc[j])
            change = sym_df['Change'].iloc[j]
            df['Total Count'].iloc[x + category] += 1
            if(change > 0):
                df['Increase Count'].iloc[x + category] += 1
            else:
                df['Decrease Count'].iloc[x + category] += 1
            df['Average Change'].iloc[x + category] += change
    x+=16
```

```python
for i in range(len(df)):
    totalCount = df['Total Count'][i]
    if(totalCount == 0):
        totalCount+=1
    df['Increase Percentage'][i] = round(df['Increase Count'][i]*100/totalCount, 2) 
    df['Decrease Percentage'][i] = round(df['Decrease Count'][i]*100/totalCount, 2)
    df['Average Change'][i] = round(df['Average Change'][i]*100/totalCount , 2)
                                         
```
