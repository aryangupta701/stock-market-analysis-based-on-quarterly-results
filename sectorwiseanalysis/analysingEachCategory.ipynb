```python
import pandas as pd
```

```python
sector_df = pd.read_csv('C:\\Users\\garya\\ML projects\\Stock analysis\\sectorwiseanalysis\\dataset\\sectorIndex.csv').set_index('SYMBOL')
```

```python
sectors = ['metal', 'financial services', 'oil gas']
```

```python
for sector in sectors:
    savePath = f"C:\\Users\\garya\\ML projects\\Stock analysis\\sectorwiseanalysis\\dataset\\{sector}.csv"
    frames = []
    for i in range(16):
        increaseCount = 0 
        decreaseCount = 0
        path = f"C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\16categories\\category_number_{i}.csv"
        df = pd.read_csv(path)
        totalCount = 0
        for j in range(len(df)):
            flag = False
            symbol = df['SYMBOL'].loc[j]
            for x in sector.split(' '):
                if(symbol in sector_df.index and (x in str(sector_df['industry'].loc[symbol]).lower())):
                    flag = True
                    break
                
            if(flag == False):
                continue
            print("heelo")
            totalCount+=1;
            if(df['Change'].loc[j] < 0):
                decreaseCount += 1
            else:
                increaseCount += 1
        avg = df['Change'].mean()
        
        if(totalCount == 0):
            totalCount = 1
        data = {
            "Category" : i, 
            "Total Count" : totalCount, 
            "Decrease Count" : decreaseCount, 
            "Increase Count" : increaseCount,
            "Decrease Percentage" : f"{round((decreaseCount*100)/totalCount,2)}%",
            "Increase Percentage" : f"{round((increaseCount*100)/totalCount,2)}%"
        }
        temp_df = pd.DataFrame(data, index = [1])
        frames.append(temp_df)
    final_df = pd.concat(frames, ignore_index=True)
    final_df.to_csv(savePath, index = False)
    
```
