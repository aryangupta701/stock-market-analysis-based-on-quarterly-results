```python
def extractNum(s):
    ans = ""
    flag = False
    for c in s:
        if(flag and c == '.'):
            ans += c
            continue
        if(c >= '0' and c<='9'):
            flag = True
        if(flag and c != '\n' and c!=' '):
            ans += c
    return ans
```
