```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\garya\\ML projects\\stock market telegram\\dataset\\previous13QuartersData.csv").set_index("Scrip_Name")
highs = pd.read_csv("C:\\Users\\garya\\ML projects\\stock market telegram\\dataset\\highdataQuarterWise.csv").set_index("ScripCodes")
scrips = pd.read_csv("C:\\Users\\garya\\ML projects\\stock market telegram\\dataset\\scrips.csv")
lows = pd.read_csv("C:\\Users\\garya\\ML projects\\stock market telegram\\dataset\\lowdataQuarterWise.csv").set_index("ScripCodes")
```

```python
data = data.dropna()
data
```

```python
highs = highs.dropna()
```

```python
lows = lows.dropna()
```

```python
marginDownCount = 0
marginUpCount = 0
totalDown = 0
totalUp = 0
quarter = lows.columns
for scrip in scrips['Scrip_Name']:
    for i in range(len(quarter)-1):
        
        if((scrip in data.index) and (scrip in highs.index ) and (scrip in lows.index)):
            try:
    #             print(scrip)
                str1 = data[f"Margin {quarter[i+1]}"][scrip]
                str2 = data[f"Margin {quarter[i]}"][scrip]
#                 print(quarter[i+1], quarter[i], str1, str2)
    #             print(str1)
    #             print(str2)
                a = float(str1[:-1])
                b = float(str2[:-1])
                change = a - b
                if(change < 0):
                    if(float(highs[quarter[i+1]][scrip]) > float(highs[quarter[i]][scrip])):
                        marginDownCount += 1
                    totalDown+=1
                else:
                    if(float(highs[quarter[i+1]][scrip]) > float(highs[quarter[i]][scrip]) ):
                        marginUpCount += 1
                    totalUp+=1
            except Exception as e:
                continue

print(f"Probability of stock going up if change in Margin is Positive -> {marginUpCount/totalUp}")
print(f"Probability of stock going up if change in Margin is Negative -> {marginDownCount/totalDown}")
```
