```python
dataHigh = {
#     "SYMBOL" : scrip_codes['ScripCodes'],
    "Sep 2019" : np.full(scrip_codes.size,0),
    "Dec 2019" : np.full(scrip_codes.size,0),
    "Mar 2020" : np.full(scrip_codes.size,0), 
    "Jun 2020" : np.full(scrip_codes.size,0),   
    "Sep 2020" : np.full(scrip_codes.size,0), 
    "Dec 2020" : np.full(scrip_codes.size,0),
    "Mar 2021" : np.full(scrip_codes.size,0), 
    "Jun 2021" : np.full(scrip_codes.size,0),   
    "Sep 2021" : np.full(scrip_codes.size,0), 
    "Dec 2021" : np.full(scrip_codes.size,0),
    "Mar 2022" : np.full(scrip_codes.size,0), 
    "Jun 2022" : np.full(scrip_codes.size,0),   
    "Sep 2022" : np.full(scrip_codes.size,0), 
}
dataLow = {
#     "SYMBOL" : scrip_codes['ScripCodes'],
    "Sep 2019" : np.full(scrip_codes.size,1000000),
    "Dec 2019" : np.full(scrip_codes.size,1000000),
    "Mar 2020" : np.full(scrip_codes.size,1000000), 
    "Jun 2020" : np.full(scrip_codes.size,1000000),   
    "Sep 2020" : np.full(scrip_codes.size,1000000), 
    "Dec 2020" : np.full(scrip_codes.size,1000000),
    "Mar 2021" : np.full(scrip_codes.size,1000000), 
    "Jun 2021" : np.full(scrip_codes.size,1000000),   
    "Sep 2021" : np.full(scrip_codes.size,1000000), 
    "Dec 2021" : np.full(scrip_codes.size,1000000),
    "Mar 2022" : np.full(scrip_codes.size,1000000), 
    "Jun 2022" : np.full(scrip_codes.size,1000000),   
    "Sep 2022" : np.full(scrip_codes.size,1000000), 
}
# print(dataHigh)
highdf = pd.DataFrame(dataHigh, index = [scrip_codes['ScripCodes']])
lowdf = pd.DataFrame(dataLow, index =[scrip_codes['ScripCodes']])
print(lowdf, highdf)
```

```python
for i in range(len(dates)): 
    try: 
        date = dates['Date'][i]
        year = str(dates['Year'][i])
        yearnum = dates['Year'][i]
        mon = dates['Month'][i]
        URL = f"https://www1.nseindia.com/content/historical/EQUITIES/{year}/{mon}/cm{date}bhav.csv.zip"
        res = requests.get(URL)
        if(res.status_code == 404):
            continue
        data = pd.read_csv(URL)
        symbols = data['SYMBOL']
        highprices = data['HIGH']
        lowprices = data['LOW']
        if(mon == 'JAN' or mon == 'FEB' or mon == 'MAR'):
            quarter = 'Dec ' + str(yearnum-1)
        elif(mon == 'APR' or mon == 'MAY' or mon == 'JUN'):
            quarter = 'Mar ' + year
        elif(mon == 'JUL' or mon == 'AUG' or mon == 'SEP'):
            quarter = 'Jun ' + year
        else:
            quarter = 'Sep ' + year
        print(quarter, date)
        for j in range(len(highprices)):
#             print(highprices[j], symbols[j])
            if((quarter in highdf.columns) and (symbols[j] in highdf.index)):
                highdf[quarter][symbols[j]] = max(highdf[quarter][symbols[j]], highprices[j])
    
            if((quarter in lowdf.columns) and (symbols[j] in lowdf.index)):
                lowdf[quarter][symbols[j]] = min(lowdf[quarter][symbols[j]], lowprices[j])       
    except Exception as e:
        print(e)
```
