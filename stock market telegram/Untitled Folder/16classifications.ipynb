```python
def getChangeVal(str1, str2, flag=False):
    
    if(flag): 
        a = float(str1[:-1])
        b = float(str2[:-1])
    else:
        a = float(str1)
        b = float(str2)
    change = a - b
    if(change > 0):
        return 1
    return 0
```
