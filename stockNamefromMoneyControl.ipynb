```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup,SoupStrainer
import numpy as np
import time
import html5lib
frames = []
```

```python
def getSymbols(link):
    try:
        
        status_code = -1
        while status_code != 200:
            h = requests.get(link)
            status_code = h.status_code
            print("symbols",status_code)
            if(status_code == 404):
                return -1,-1
            if(status_code == 429):
                print("symbol sleeping")
                time.sleep(5)
            elif(status_code != 200):
                return -1,-1

        soup1 = BeautifulSoup(h.text,"lxml")
        sym = soup1.findAll('li', class_="clearfix")
        nsesymbol = ""
        bsesymbol = ""
        flag1 = False 
        flag2 = False
        for li in sym:
            for con in li.contents:
                val = con.string
    #             print(val,"yo")
                if(val == None):
                    continue
                if(val == "Series"):
                    return nsesymbol, bsesymbol
                elif(con.string == "BSE:"):
                    flag1 = True
                    continue
                elif(val == "NSE:"):
                    flag1 = False
                    flag2 = True
                    continue
                if(flag1 and len(val) >= 2):
                    bsesymbol = val
                    flag1 = False
                if(flag2 and len(val) >= 2):
                    nsesymbol = val 
                    return nsesymbol, bsesymbol
    except Exception as e:
        print(e)
        return -1, -1
    # return nsesymbol, bsesymbol
```

```python
for i in range(0,27):
    if( i == 26):
        page = "others"
    else:
        page = chr(65+i)
    status_code = -1
    URL = f"https://www.moneycontrol.com/india/stockpricequote/{page}"
    while status_code != 200:
        html = requests.get(URL)
        status_code = html.status_code
        print(status_code)
        if(status_code == 404):
            break
        if(status_code == 429):
            print("sleeping...")
            time.sleep(5)
        elif(status_code != 200):
            break
    soup = BeautifulSoup(html.text, parseOnlyThese=SoupStrainer(class_ = "bl_12"))
    atags = soup.findAll("a")
    print(URL)
    for td in atags:
        try: 
            link = td.get("href")                
            list1 = link.split("/")
            nsesymbol, bsesymbol = getSymbols(link)
            if(nsesymbol == -1):
                continue
            data = {
                "URL" : link,
                "companysector" : list1[-3],
                "companyname" : list1[-2],
                "companycode" : list1[-1],
                "nsesymbol" : nsesymbol,
                "bsesymbol" : bsesymbol
            }
            temp_df = pd.DataFrame(data,index=[1])
            frames.append(temp_df)
            print(list1[-2])
        except Exception as e:
            print(e)
```

```python
data_df = pd.concat(frames, ignore_index=True)
```
