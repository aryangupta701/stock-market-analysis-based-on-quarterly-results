```python
import datetime
import pandas as pd
from dateutil.relativedelta import relativedelta
import os
import math
```

```python
def getQuarters(curDate, prevDate):
    
    prevMon = prevDate.strftime("%b")
    curMon = curDate.strftime("%b")
    prevYear = prevDate.strftime("%Y")
    curYear = curDate.strftime("%Y")
    curQuarter = f"{curMon} {curYear}"
    prevQuarter = f"{prevMon} {prevYear}"
    return curQuarter, prevQuarter

def quarterList():
    lst = []
    date = datetime.datetime(2022,12,1)
    minDate = datetime.datetime(2017,1,1)
    while(date > minDate):
        prevDate = date-relativedelta(months=3)
        curQuarter, prevQuarter = getQuarters(date, prevDate)
        lst.append([curQuarter, prevQuarter])
        date = prevDate
    return lst
```

```python
#   0         0          0         0
#   8         4          2         1
#(profit) (revenue)  (Ebitda) `(margin)

def getBit(str1, str2):
#     print(str1,str2)
#     print(str1, str2)
    if(math.isnan(str1)  or math.isnan(str2)):
        return 20, -1
    a = float(str1)
    b = float(str2)
    change = (a - b)/a
    #change = (curQuartervalue - prevQuartervalue) / curQuarterValue
#     print(a,b)
    if(change > 0):
        return 1,change
    return 0,change

def generateCategory(prevQuarter, curQuarter, fundamental_row):
    try:
        ans = 0
        marginBit,marginChange = getBit(fundamental_row[f"Margin {curQuarter}"], fundamental_row[f"Margin {prevQuarter}"])
        ebitdaBit,ebitdaChange = getBit(fundamental_row[f"EBITDA {curQuarter}"], fundamental_row[f"EBITDA {prevQuarter}"])
        revenueBit,revenueChange = getBit(fundamental_row[f"Revenue {curQuarter}"], fundamental_row[f"Revenue {prevQuarter}"])
        profitBit,profitChange = getBit(fundamental_row[f"Profit {curQuarter}"], fundamental_row[f"Profit {prevQuarter}"])

        ans += marginBit*1
        ans += ebitdaBit*2 
        ans += revenueBit*4
        ans += profitBit*8
        
        if(ans > 15):
            return -10,-10,-10,-10
        return marginChange, ebitdaChange, revenueChange, profitChange
    except Exception as e:
        print("category", e)
        return -10,-10,-10,-10

def getCurPrice(date, symbol, flag=True):
    if(flag):
        date = date.split("-")
        if(len(date) == 1):
            date = date[0].split("/")
        date = datetime.datetime(int(date[2]),int(date[1]),int(date[0]) )
        
    minDate = datetime.datetime(2017,1,1)
    while(date > minDate ):
        try:
            YY = date.strftime("%Y")
            mon = date.strftime("%b").upper()
            dd = date.strftime("%d")
            df = pd.read_csv(f"C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\nsedataPrevious5years\\cm{dd}{mon}{YY}bhav.csv").set_index("SYMBOL")
            if(symbol not in df.index):
                return -1
            ans = df['CLOSE'][symbol]
            return ans
        except Exception as e:
            x = 1           
#           print(e)
                
        finally:
            date -= datetime.timedelta(days=1)
    return -1

def getHighLowValues(date, symbol):
#     print(symbol)
    high = getCurPrice(date,symbol)
    low = high
    date = date.split("-")
    if(len(date) == 1):
        date = date[0].split("/")
    date = datetime.datetime(int(date[2]),int(date[1]),int(date[0]) )
    for i in range(1,90):
        date += datetime.timedelta(days=1)
        val = float(getCurPrice(date,symbol,False))
        if(val == -1):
            continue
        high = max(high,val )
        low = min(low, val)
    
    return high,low
        
```

```python
def predictdate(curQuarter):
    mon = curQuarter.split(" ")[0]
    year = curQuarter.split(" ")[1]
    if(mon == "Mar"):
        return f"30-06-{year}"
    elif(mon == "Jun"):
        return f"14-08-{year}"
    elif(mon == "Dec"):
        return f"13-02-{int(year)+1}"
    else:
        return f"12-11-{year}"
```

```python
frames = []
for i in range(len(quarter_result_dates)):
# for i in range(6,7):
    print(i)
    try: 
        date_row = quarter_result_dates.iloc[i]
#         print(date_row)
        symbol = date_row['nsesymbol']
        companyname = date_row['companyname']
        companycode = date_row['companycode']
        columns = date_row.index
        fundamental_row = quarterly_df.loc[symbol]
#         print(fundamental_row)
        for qtr in quarterPairs:
            try:
                curQuarter = qtr[0]
                prevQuarter = qtr[1]
                
                if((curQuarter in columns)):
                    margin,ebitda,revenue,profit = generateCategory(prevQuarter, curQuarter, fundamental_row)
                    ddate = date_row[curQuarter]
                    if(ddate == ""):
                        ddate = predictdate(curQuarter)
                    if(margin == -10):
                        continue
                    print(symbol, curQuarter, prevQuarter)
                    highValue , lowValue = getHighLowValues(ddate, symbol)
#                     print(highValue, lowValue)
                    curPrice = getCurPrice(ddate, symbol)
                    if(curPrice == -1):
                        continue
                    highChange = (highValue-curPrice)/curPrice
                    lowChange = (lowValue-curPrice)/curPrice

                    if(abs(highChange) > abs(lowChange)):
                        change = highChange
                    else:
                        change = lowChange

                data = {
                    "SYMBOL" : symbol,
                    "Company Name" : companyname,
                    "Company Code" : companycode,
                    "Quarter" : f"{prevQuarter} - {curQuarter}",
                    "Result Date": date_row[curQuarter],
                    "Price on Result Date": curPrice,
                    "High in next 90 days" : highValue,
                    "Low in next 90 days" : lowValue, 
                    "margin": margin,
                    "ebitda": ebitda,
                    "revenue": revenue,
                    "profit":profit,
                    "Change" : change, 
                }
                temp_df = pd.DataFrame(data, index=[1])
                frames.append(temp_df)
            except Exception as e:
                print("For", e)
            
    except Exception as e:
        print("main", e)
```

```python
final_df = pd.concat(frames, ignore_index = True)
```
