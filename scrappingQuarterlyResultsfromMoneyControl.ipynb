```python
frames = []
for i in range(len(company_data)):
    try:
        companyname = company_data['companyname'][i]
        companycode = company_data['companycode'][i]
        symbol = company_data['symbol'][i]
        if(len(symbol) == 0):
            continue
        print(f"https://www.moneycontrol.com/financials/{companyname}/results/consolidated-quarterly-results/{companycode}#{companycode}")
        data = []
        for pagenumber in range(1,5):
            try:
                URL = f"https://www.moneycontrol.com/financials/{companyname}/results/consolidated-quarterly-results/{companycode}/{pagenumber}#{companycode}"
                status_code = -1
                while status_code != 200:
                    html = requests.get(URL)
                    status_code = html.status_code
                    print(status_code)
                    if(status_code == 404):
                        break
                    if(status_code != 200):
                        print("Sleeping..")
                        time.sleep(5)
                soup = BeautifulSoup(html.text, "lxml")
                table_html = soup.find(class_ = "mctable1")
                temp_df = pd.read_html(str(table_html))[0]
                temp_df = (temp_df.iloc[1:]
                      .set_axis(temp_df.iloc[0], axis=1)
                      .set_index(temp_df.iat[0,0])
                      .rename_axis(index=None, columns=None))
                data.append(temp_df.transpose())
            except Exception as e:
                print(e)

        results_df = pd.concat(data)
        dict1 = {
            "symbol" : symbol,
            "companyname" : companyname,
            "companycode" : companycode
        }
#         print(results_df)
        indexs = results_df.index
        indexs = indexs.dropna()
        for mon in indexs:
            if(mon == None):
                continue
            month = mon.replace("'", "20")
            dict1.__setitem__(f"Revenue {month}", results_df['Net Sales/Income from operations'][mon])
#             dict1.__setitem__(f"EBITDA {month}", results_df['P/L Before Int., Excpt. Items & Tax'][mon])    
            dict1.__setitem__(f"EBITDA {month}", results_df['P/L Before Other Inc., Int., Excpt. Items & Tax'][mon])    
            
            dict1.__setitem__(f"Profit {month}", results_df['Net Profit/(Loss) For the Period'][mon])    
#             dict1.__setitem__(f"Margin {month}", round(float(results_df['P/L Before Int., Excpt. Items & Tax'][mon])*100 / float(results_df['Net Sales/Income from operations'][mon]) , 2)) 
            dict1.__setitem__(f"Margin {month}", round(float(results_df['P/L Before Other Inc., Int., Excpt. Items & Tax'][mon])*100 / float(results_df['Net Sales/Income from operations'][mon]) , 2)) 
    
        temp_df = pd.DataFrame(dict1,index=[1])
        frames.append(temp_df)
    except Exception as e:
        print("Main", e)
```

```python
df = pd.concat(frames,ignore_index = True)
```
