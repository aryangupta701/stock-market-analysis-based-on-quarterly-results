```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import html5lib
```

```python
company_data = pd.read_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\moneyControlListWithSymbols.csv")
company_data.head(5)
```

```python
def getMon(m):
    if(m>=2 and m<=4):
        return "Mar"
    elif(m>=5 and m<=7):
        return "Jun"
    elif(m>=8 and m<=10):
        return "Sep"
    return "Dec"
    
def getYear(m,y):
    if(m == 1):
        return y-1
    return y
frame = []
```

```python

for i in range(len(company_data)):
    try:
        companyname = company_data['companyname'][i]
        companycode = company_data['companycode'][i]
        URL = f"https://www.moneycontrol.com/company-facts/{companyname}/board-meetings/{companycode}#{companycode}"
        print(URL)
        html_text = requests.get(URL)
        soup = BeautifulSoup(html_text.text, 'lxml')
        dateData = soup.find(class_ = "board-meeting-table")
        data_df = pd.read_html(str(dateData))[0]
        data = {
            "companyname" : companyname,
            "companycode" : companycode,
            "nsesymbol" : company_data['nsesymbol'][i],
            "bsesymbol" : company_data['bsesymbol'][i]
        }
        for i in range(len(data_df)):
            try:
                list1 = data_df['Remark'][i].lower().split(' ')
                if("results" not in list1):
                    continue

                date = data_df['Meeting Date'][i].split('-')
                mon = getMon(int(date[1]))
                year = getYear(int(date[1]),int(date[2]))

                if(data.get(f"{mon} {year}", "Not found") != "Not found"):
                    continue

                data.__setitem__(f"{mon} {year}", data_df['Meeting Date'][i])
            except Exception as e:
                print("Inside For : " ,e)

        temp_df = pd.DataFrame(data, index=[1])
        frame.append(temp_df)
    except Exception as e:
        print(e)
```

```python
df = pd.concat(frame, ignore_index = True)
```
