```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup,SoupStrainer
import numpy as np
import time
import html5lib
company_data = pd.read_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\tempdata\\quarterlyResultsNSE.csv")
company_data = company_data.fillna("")
```

```python
frames = []
for i in range(len(mclist)):
# for i in range(1):
    try:
        companyname = mclist['companyname'][i]
        companycode = mclist['companycode'][i]
        symbol = mclist['nsesymbol'][i]
        print(symbol)
        if(symbol not in lst):
            continue
        if(len(symbol) == 0):
            continue
        print(f"https://www.moneycontrol.com/financials/{companyname}/results/consolidated-quarterly-results/{companycode}#{companycode}")
        data = []
        for pagenumber in range(1,5):
            try:
                URL = f"https://www.moneycontrol.com/financials/{companyname}/results/consolidated-quarterly-results/{companycode}/{pagenumber}#{companycode}"
                status_code = -1
                while status_code != 200:
                    html = requests.get(URL)
                    status_code = html.status_code
                    print(status_code)
                    if(status_code == 404):
                        break
                    if(status_code != 200):
                        print("Sleeping..")
                        time.sleep(5)
                soup = BeautifulSoup(html.text, "lxml")
                table_html = soup.find(class_ = "mctable1")
                temp_df = pd.read_html(str(table_html))[0]
                temp_df = (temp_df.iloc[1:]
                      .set_axis(temp_df.iloc[0], axis=1)
                      .set_index(temp_df.iat[0,0])
                      .rename_axis(index=None, columns=None))
                data.append(temp_df.transpose())
            except Exception as e:
                print(e)

        results_df = pd.concat(data)
        dict1 = {
            "symbol" : symbol,
            "companyname" : companyname,
            "companycode" : companycode
            }
#         print(results_df)
        indexs = results_df.index
        indexs = indexs.dropna()
        for mon in indexs:
            if(mon == None):
                continue
            month = mon.replace("'", "20")
            rev = (float(results_df['(a) Int. /Disc. on Adv/Bills'][mon]) + 
                    float(results_df['(b) Income on Investment'][mon]) + 
                    float(results_df['(c) Int. on balances With RBI'][mon]) + 
                    float(results_df['(d) Others'][mon])) 
            ebit = (float(results_df['Operating Profit before Provisions and contingencies'][mon]) - 
                        float(results_df['Provisions And Contingencies'][mon]) )
            dict1.__setitem__(f"Revenue {month}", rev)
            dict1.__setitem__(f"EBITDA {month}", ebit)    
            dict1.__setitem__(f"Profit {month}", float(results_df['Net Profit/(Loss) For the Period'][mon]) + float(results_df['Share Of P/L Of Associates'][mon]) )   
            dict1.__setitem__(f"Margin {month}", round(ebit*100 / rev , 2) )
            
        temp_df = pd.DataFrame(dict1,index=[1])
        frames.append(temp_df)
    except Exception as e:
        print("Main", e)
```

```python
df = pd.concat(frames, ignore_index = True)
frames
```
