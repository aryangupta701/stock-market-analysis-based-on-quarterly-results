```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\modelling\\dataset\\categoryWithChange.csv")
quarterly_res = pd.read_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\datasets\\quarterlyResultsNSE.csv").set_index("symbol")
```

```python
df
```

```python
frames = []
for i in range(len(df)):
    try:
        quarters = df['Quarter'][i].split(' -')
        symbol = df['SYMBOL'].loc[i]
        print(symbol)
        curRevenue = float(quarterly_res[f"Revenue{quarters[1]}"].loc[symbol])
        prevRevenue = float(quarterly_res[f"Revenue {quarters[0]}"].loc[symbol])

        curEbitda = float(quarterly_res[f"EBITDA{quarters[1]}"].loc[symbol])
        prevEbitda = float(quarterly_res[f"EBITDA {quarters[0]}"].loc[symbol])

        curProfit = float(quarterly_res[f"Profit{quarters[1]}"].loc[symbol])
        prevProfit = float(quarterly_res[f"Profit {quarters[0]}"].loc[symbol])

        curMargin = float(quarterly_res[f"Margin{quarters[1]}"].loc[symbol])
        prevMargin = float(quarterly_res[f"Margin {quarters[0]}"].loc[symbol])

        revenueIncrease = round( (curRevenue - prevRevenue) / prevRevenue, 2 ) 
        ebitdaIncrease = round( (curEbitda - prevEbitda) / prevEbitda , 2)
        profitIncrease = round( (curProfit - prevProfit) / prevProfit ,2 )
        marginIncrease = round( (curMargin - prevMargin) / prevMargin , 2)

        print(revenueIncrease , ebitdaIncrease , profitIncrease , marginIncrease , symbol)
        data = {
            "SYMBOL" : symbol, 
            "Category" : df['category'][0],
            "Quarter" : df['Quarter'][i],
            "RevenueChange" : revenueIncrease , 
            "EBITDAChange" : ebitdaIncrease, 
            "ProfitChange" : profitIncrease , 
            "MarginChange" : marginIncrease, 
            "Change" : df['Change'][i]
        }
        temp_df = pd.DataFrame(data, index = [1])
        frames.append(temp_df)
    except Exception as e:
        print(e)
```

```python
final_df = pd.concat(frames, ignore_index = True)
```

```python
final_df.to_csv("C:\\Users\\garya\\ML projects\\Stock analysis\\modelling\\dataset\\trainDataset.csv", index = False)
```
